#  都内でおすすめのカフェ5選

```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
JonesCafeBar = [35.685689958293516, 139.69210725235016]
jaho = [35.646256019537944, 139.74503494807047]
NINAS = [35.67431000616222, 139.7600507693085]
Buvette = [35.67391352619888, 139.75919765420448]
accueil = [35.651040103534676, 139.7069992233782]
nogizaka = [35.666961675658605, 139.72650782519426]


# make the map
m = folium.Map(location=nogizaka, zoom_start=12)

# add a marker
folium.Marker(
    JonesCafeBar, 
    popup="The Jones Cafe Bar, the best place on earth!", 
    tooltip="The Jones Cafe Bar",
    icon=folium.Icon(color="lightblue", icon="cutlery")
).add_to(m)

folium.Marker(
    jaho, 
    popup="Jahocoffeejapan, the best place on earth!", 
    tooltip="Jahocoffeejapan",
    icon=folium.Icon(color="pink", icon="cutlery")
).add_to(m)

folium.Marker(
    NINAS, 
    popup="LE SALON DE NINA'S, the best place on earth!", 
    tooltip="LE SALON DE NINA'S",
    icon=folium.Icon(color="beige", icon="cutlery")
).add_to(m)

folium.Marker(
    Buvette, 
    popup="Buvette Tokyo, the best place on earth!", 
    tooltip="Buvette Tokyo",
    icon=folium.Icon(color="purple", icon="cutlery")
).add_to(m)

folium.Marker(
    accueil, 
    popup="cafe accueil, the best place on earth!", 
    tooltip="cafe accueil",
    icon=folium.Icon(color="red", icon="cutlery")
).add_to(m)

# show the map
m
```

```python
m.save('Tokyo_cafe.html')
```
