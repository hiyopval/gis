# 茨城県の地図

```python
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
ibaraki = gpd.read_file('ibaraki.geojson')
```

```python
ibaraki.plot()
```

```python
ibaraki_pop = pd.read_csv('ibaraki.csv')
```

```python
# rename
ibaraki.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
ibaraki
```

```python
# OBJECTからintに変える
ibaraki['KEY_CODE'] = ibaraki['KEY_CODE'].astype(int)
```

```python
ibaraki.info()
```

```python
# ここでようやく merge コマンドで統合させて、新しい変数に入れる
ibaraki_pop = ibaraki.merge(ibaraki_pop,on='KEY_CODE')
```

```python
# check!
ibaraki_pop
```

```python
ibaraki_pop.info()
```

```python
# change from OBJECT to int
ibaraki_pop['人口総数'] = ibaraki_pop['人口総数'].astype(int)
```

```python
# check
ibaraki_pop.info()
```

```python
ibaraki_pop.plot(figsize=(10,10),
                  column='人口総数',
                  legend=True,
                  cmap='RdPu')
```

```python
# 市区町村名の表示
# Centroidの計算
ibaraki_pop['centroid'] = ibaraki_pop.geometry.centroid
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# 地図のプロット
fig, ax = plt.subplots(1, 1, figsize=(8, 8))  # サイズを適度に設定
ibaraki_pop.plot(column='人口総数', ax=ax, legend=True, cmap='RdPu')

# 市区町村名の表示
# Centroidの計算
ibaraki_pop['centroid'] = ibaraki_pop.geometry.centroid

# テキストの追加
for idx, row in ibaraki_pop.iterrows():
    ax.text(row['centroid'].x, row['centroid'].y, row['CITYNAME'], fontsize=8, ha='center')

plt.title('茨城県の市区町村別人口総数')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 日本語フォントの設定（例: WindowsのMS Gothic）
fp = FontProperties(fname='C:\\Windows\\Fonts\\msgothic.ttc', size=8)  # WindowsのMS Gothic
# fp = FontProperties(fname='/System/Library/Fonts/ヒラギノ角ゴシック W6.ttc', size=8)  # macOSの例
# fp = FontProperties(fname='/usr/share/fonts/truetype/noto/NotoSansCJK-Regular.ttc', size=8)  # Linuxの例

# 地図のプロット
fig, ax = plt.subplots(1, 1, figsize=(8, 8))  # サイズを適度に設定
ibaraki_pop.plot(column='人口総数', ax=ax, legend=True, cmap='RdPu')

# 市区町村名の表示
# Centroidの計算
ibaraki_pop['centroid'] = ibaraki_pop.geometry.centroid

# 一意の市区町村名とその位置を取得
unique_labels = ibaraki_pop.drop_duplicates(subset=['CITYNAME'])

# テキストの追加
for idx, row in unique_labels.iterrows():
    ax.text(row['centroid'].x, row['centroid'].y, row['CITYNAME'], fontsize=8, ha='center', fontproperties=fp)

plt.title('茨城県の市区町村別人口総数', fontproperties=fp)
plt.show()
```
