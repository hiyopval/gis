# 吹奏楽コンクール全国大会に出場した学校マップ（過去5年間、東関東地区のみ）


```python
# library import
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wind.csv')
```

```python
df
```

```python
df.info()
```

```python
df.plot.scatter(x='緯度',y='経度',s=50)
```

```python
# complete the following code:
center_lat = df['緯度'].mean()
center_lon = df['経度'].mean()
print(center_lat)
print(center_lon)
```

```python
import folium

# 地図を作成
m = folium.Map(location=[center_lat, center_lon], zoom_start=8)

# DataFrame をループしてマーカーを作成
for index, row in df.iterrows():
    latlon = [row['緯度'], row['経度']]
    folium.Marker(
        latlon, 
        popup='<div style="width:350px">' + row['賞名'] + '</div>', 
        tooltip=row['学校名']
    ).add_to(m)

# 地図を表示
m
```

```python
m.save("map.html")
```
